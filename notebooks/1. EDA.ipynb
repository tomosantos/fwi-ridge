## Algerian Forest Fires Dataset


### Data Set Information:

The dataset includes 244 instances collected from two regions of Algeria: Bejaia region in the northeast and Sidi Bel-abbes region in the northwest.

- 122 instances for each region.
- The period covers June 2012 to September 2012.
- The dataset contains 11 attributes and 1 output attribute (class).
- The 244 instances are classified into fire (138 classes) and non-fire (106 classes).

**Attribute Information:**

1. Date: (DD/MM/YYYY) Day, month ('June' to 'September'), year (2012) - Meteorological observations
2. Temp: Temperature at noon (maximum temperature) in Celsius: 22 to 42
3. RH: Relative Humidity in %: 21 to 90
4. Ws: Wind speed in km/h: 6 to 29
5. Rain: Total daily rainfall in mm: 0 to 16.8
6. FFMC: Fine Fuel Moisture Code from the FWI system: 28.6 to 92.5
7. DMC: Duff Moisture Code from the FWI system: 1.1 to 65.9
8. DC: Drought Code from the FWI system: 7 to 220.4
9. ISI: Initial Spread Index from the FWI system: 0 to 18.5
10. BUI: Buildup Index from the FWI system: 1.1 to 68
11. FWI: Fire Weather Index: 0 to 31.1
12. Classes: Two classes, namely Fire and Not Fire

## 1. Data Cleaning

```python
## Missing Values
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df.loc[:122, 'Region'] = 0
df.loc[122:, 'Region'] = 1

df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
## remove the 122sd row
df = df.drop(122).reset_index(drop=True)
df.iloc[[122]]
```

```python
df.columns
```

```python
# Fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

### Change the required columns as integers data type

```python
df.columns
```

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)

df.info()
```

### Changing the other columns to float data

```python
objects = [features for features in df.columns if df[features].dtypes == 'O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)

df.info()
```

```python
df.describe()
```

```python
## Saving the cleaned dataset
os.makedirs('data', exist_ok=True)

df.to_csv('data/AlgerianForestFires_cleaned.csv', index=False)
```
